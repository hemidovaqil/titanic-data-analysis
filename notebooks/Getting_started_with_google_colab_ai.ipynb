```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
csv_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(csv_url)
```

```python
display(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

### Анализ возраста и выживаемости

```python
# Заполнение пропущенных значений в 'Age' медианой
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
print(f"Пропущенные значения в 'Age' заполнены медианой: {median_age:.1f}")

# Проверка на наличие пропущенных значений после заполнения
print("\nПропущенные значения после заполнения 'Age':")
display(df.isnull().sum())
```

```python
# Визуализация распределения возраста для выживших и невыживших
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Распределение возраста для выживших и невыживших пассажиров')
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.legend(title='Выжил', labels=['Нет', 'Да'])
plt.show()
```
